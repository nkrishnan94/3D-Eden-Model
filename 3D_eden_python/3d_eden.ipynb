```python
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)


def randrange(n, vmin, vmax):
    '''
    Helper function to make an array of random numbers having shape (n, )
    with each number distributed Uniform(vmin, vmax).
    '''
    return (vmax - vmin)*np.random.rand(n) + vmin

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

n = 100

# For each set of style and range settings, plot n random points in the box
# defined by x in [23, 32], y in [0, 100], z in [zlow, zhigh].
for m, zlow, zhigh in [('o', -50, -25), ('^', -30, -5)]:
    xs = randrange(n, 23, 32)
    ys = randrange(n, 0, 100)
    zs = randrange(n, zlow, zhigh)
    ax.scatter(xs, ys, zs, marker=m)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
size= 10
state = np.zeros((10,10))

state[5,5]=1
neighb = [[-1,0],[1,0],[0,1],[0,-1], ]
```

```python

for t in range(10):
    cell_list = np.argwhere(state==1)
    np.random.shuffle(cell_list)
    for i in range(len(cell_list)):
        filled= False
        n=0
        np.random.shuffle(neighb)
        while not filled and n <4:
            if np.all(neighb[n] + cell_list[i]<10) and np.all(neighb[n] + cell_list[i]>-1):
                if state[tuple(neighb[n] + cell_list[i])]==0:
                    state[tuple(neighb[n] + cell_list[i])]=1
                    filled =True

            n+=1

    
```
